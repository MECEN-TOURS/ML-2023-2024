```python
@dataclass(frozen=True, slots=True)
class Ligne:
    symbole: str
    lien_nyse: str
    nom: str
    page_wiki: str
    secteur: str
    siege_social: str
    introduction_indice: str
    code_cik: str
    fondation: str
    
```

```python
lignes[-1]
```

```python
motif_colonne = re.compile("<td>(.*?)</td>")
```

```python
motif_colonne.findall(lignes[-1])
```

```python
motif_sy = re.compile("""<a.*?href="(.*?)">(.*?)</a>""")
motif_n = re.compile("""<a href="(.*?)">(.*?)</a>""")
motif_ss = re.compile("<a.*?>(.*?)</a>")
```

```python
def analyse_ligne(ligne: str) -> Ligne:
    csy, cn, secteur, _, css, introduction_indice, code_cik, fondation = motif_colonne.findall(ligne)
    symbole, lien_nyse = motif_sy.search(csy).groups()
    page_wiki, nom = motif_n.search(cn).groups()
    siege_social, = motif_ss.search(css).groups()
    return Ligne(
        symbole=symbole,
        lien_nyse=lien_nyse,
        nom=nom,
        secteur=secteur,
        page_wiki=page_wiki,
        introduction_indice=introduction_indice,
        siege_social=siege_social,
        code_cik=code_cik,
        fondation=fondation
    )
```
